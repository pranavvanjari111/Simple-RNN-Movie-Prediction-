```python
from keras.src.legacy.preprocessing.text import one_hot
```

```python


        sent = ['the glass of milk',
                'the glass of juice',
                'the cup of tea',
                'I am good boy',
                'I am good developer',
                'understand the meaning of words',
                ' your videos are good']

        sent
```

```python
#define the vocabulary size
voc_size = 10000
```

```python
#one hot representations
one_hot_reper= [one_hot(word, voc_size) for word in sent]
print(one_hot_reper)
```

```python
# word Emdedding Representation

from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8  # max length

embeded_doc = pad_sequences(one_hot_reper, padding='pre',maxlen=sent_length)
print(embeded_doc)


""" in one hot we didn't get the same dimensions, so we can you pad_sequences to make the equal dimensions"""
```

```python
#feature representation
dim=10
```

```python
model =Sequential()
model.add(Embedding(voc_size, dim, input_length = sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embeded_doc)
```
