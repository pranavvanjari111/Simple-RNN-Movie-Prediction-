```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model

```

```python
# Load the word index
word_index= imdb.get_word_index()
reverse_word_index = {value : key for key, value in word_index.items()}
```

```python
# Load the pre-trained model with relu activation

model = load_model('simple_rnn_imdb.h5')
model.summary()
```

```python
model.get_weights()
```

```python
# helper functions

def decode_review(encded_review):
    return ' '.join([reverse_word_index.get(i-3, '?') for i in encded_review])


# preprocessing the user input
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2)+3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
## prediction function

def predict_sentiment(review):
    processed = preprocess_text(review)
    prediction = model.predict(processed)

    sentiment = 'Positive' if prediction[0][0] > 0.5 else "Negative"

    return sentiment, prediction[0][0]
```

```python
# Execution

example_review = 'The movie was fantastic! The acting was great and the plot was thrilling'

print(predict_sentiment(example_review))
```

```python
print(predict_sentiment("This movie is excellent and amazing"))
print(predict_sentiment("This movie is terrible and boring"))
print(predict_sentiment("The movie was fantastic and thrilling"))
print(predict_sentiment("I hated this movie. It was awful"))
```
